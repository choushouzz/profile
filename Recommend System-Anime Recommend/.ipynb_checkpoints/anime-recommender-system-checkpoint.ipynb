## Clustering

### Preprocessing for clustering

```python
# Check dtypes, need to make sure all data are numeric(int or float) before clustering
anime_list.dtypes
```

```python
### Remove ids or string that will not do well in clustering 
## Categories values are ordinal also not helful and will be removed
## Only keep features that can be converted into numeric and contribute useful information
features = ['Score',   'Episodes', 'Aired',  
         'Duration', 'Rating',  'Popularity',
       'Members', 'Favorites', 'Watching', 'Completed', 'On-Hold', 'Dropped',
       'Plan to Watch', 'Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6',
       'Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1']

anime_list_clu = anime_all_data[features]

## Need to convert all dtype to int to apply clustering
anime_list_clu.dtypes
```

```python
# Create functions to convert all types to numeric values
Nums = ['Episodes','Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6','Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1']
def features_process(df_anime):
    df  = df_anime.copy()
    features = df.columns 
    for feature in features:
        ## If already int, ignore
        if df[feature].dtype == int:
            continue 
        ## The the type is supposed to be int but has not been converted into int, fillin missing values and convert 
        if feature in Nums:
            df[feature].fillna(value = 0, inplace = True)
            df[feature].replace('Unknown',0,inplace = True)
            df[feature] = df[feature].apply(lambda x:round(float(x)))
            df[feature] = df[feature].astype("int")
        ## The the feature is score, fillin missing value and convert into float
        elif feature == "Score":
            rated_animes = df[df[feature] != "Unknown" ][feature]
            rated_animes = rated_animes.astype("float")
            mean = rated_animes.mean()
            print(mean)
            df[feature].fillna(value = mean, inplace = True)
            df[feature].replace('Unknown',mean,inplace = True)
            df[feature] = df[feature].astype("float")
        ## The the feature is duration, convert string into number of minutes
        elif feature == "Duration":
            def duration_to_int(x):
                if x == "Unknown":
                    return 24
                tokens = x.split(" ")
                total = 0
                for i in range(len(tokens)):
                    token = tokens[i]
                    if len(tokens[i]) >= 3 and token[:3] == "min":
                        total += int(tokens[i-1])
                    elif len(tokens[i]) >= 2 and tokens[:2] == "hr":
                        total += int(tokens[i-1]) * 60
                return total
            df[feature] = df[feature].apply(duration_to_int)
            df[feature] = df[feature].astype("int") 
        ## The the feature is aired, convert string into year of starting broadcasting
        elif feature == "Aired":
            def year(x):
                if not x or x == "Unknown" :
                    return 2010
                
                x = str(x)
                x = x.split(" ")
                for item in x:
                    if len(item) == 4 and item.isdigit():
                        return int(item)
            df[feature] = df[feature].apply(year)
            df[feature] = df[feature].astype("int") 
        ## The the feature is duration, convert string into number of minutes
        elif feature == "Rating":
            levels = {'R - 17+ (violence & profanity)':17, 'PG-13 - Teens 13 or older':13, 'PG - Children': 6,
                      'G - All Ages':3,'Rx - Hentai':19,'R+ - Mild Nudity':16,'Unknown':0}
            df[feature] = df[feature].apply(lambda x: levels[x])
            df[feature] = df[feature].astype("int") 
            
    return df
```

```python
# Apply weight rating formula
def weighted_rating(df, m, C):
    v = df["rating"]['count']
    R = df["rating"]['mean']
   
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
## Create similarities matrix
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(anime_list_nlp['Genres'] + anime_list_nlp['sypnopsis'])
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Content based recommendation based on cosine similarities scores
def get_recommendations(title,df = anime_list_nlp , cosine_sim = cosine_sim):
    idx = df[df["Name"] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    i = 0
    anime_indices = []
    count = 0
    while len(anime_indices)  < 10:
        sim_score = sim_scores[i]
        index = sim_score[0]
        if index != idx:
            anime_indices.append(index)
            count += 1
        i += 1
        
    return df.iloc[anime_indices]
```

###  2nd Similarity Matrix Anime-Anime

```python
# Function: calculate the similarity, list 1 = user's list, list 2 = anime database 
# AlgorithmCosine Similarity 

def calCosineSimilarity(list1, list2): 
    res = 0
    denominator1 = 0.0
    denominator2 = 0.0
    for i in range(len(list1)):
        val1 = list1[i]
        val2 = list2[i]
        if float(val1) == 0.0 or float(val2) == 0.0:
            continue
        res += (val1 * val2)
        denominator1 += val1 ** 2
        denominator2 += val2 ** 2
    if denominator1 == 0.0 or denominator2 == 0.0:  # when denominator = 0, similarity = 0
        return 0
    else:
        return res / (math.sqrt(denominator1 * denominator2))
```

###  Recommend

```python
# if user is in the databasethen recommend based on the content similarityif not then give recomment Top 10
def recommend(user_id):
    if user_id in user_id_list:
        print('Have rated other animations, will use CF recommendation')
        
        # retrieve all the anime ID that he/she has rated and the score  3
        a_list = df_rate[(df_rate['user_id'] == user_id) & 
                                     (df_rate['rating'] >= 3)]['anime_id']
        print('Anime amount which has been rated and the score is higher than 3', len(a_list))
        
        if len(a_list) == 0: 
        # which means he/she is not interested in the anime watched before, then recommend from popularity lsit
            final_recommend_list = list()
            print('Have no interested animes before, will recommend from popular list:')
            for i in sorted(rank_anime_mapping, key=lambda x:x[1], reverse=True):
                ani_id = i[0]
                if ani_id not in a_list and ani_id not in final_recommend_list:
                    final_recommend_list.append(ani_id)
                if len(final_recommend_list) >= 10:
                    break
                    
        else:   
            recommend_list = list()
            # caculate anime's similarity one by one
            for a in a_list:
            # loop each anime ID in the watched & rated list 
                a = int(a)
                index = list_anime.index(a)
                start_list = anime_values[int(index)]
                rate_list = list()
                
                for j in range(len(anime_values)):
                    if j == index: 
                    # the anime ID is itselfthen do not caculate similarity
                        continue
                        
                    if float(animeSimMatrix[index, j]) > 0.0:
                    # which means already existed in the similarity matrix                     
                        cal1 = animeSimMatrix[index, j]
                        
                    else:
                        cal1 = calCosineSimilarity(start_list, anime_values[j])
                    # caculate the similarity between user's watched anime and other animes                     
                        animeSimMatrix[index, j] = cal1 
                        # BTW, improve the similarity matrix
                        animeSimMatrix[j, index] = animeSimMatrix[index, j]
                    rate_list.append([j, cal1])

                # from user's similarity listfind top 10.
                top10_a_list = list(sorted(rate_list, key=lambda x:x[1], reverse=True)[0:10])
                recommend_list += top10_a_list
  
            final_recommend_list = list()
            print("Find 10 most similar animations ")
            for i in sorted(recommend_list, key=lambda x:x[1], reverse=True):
                id_ = i[0]
                ani_id = list_anime[id_]
                if ani_id not in final_recommend_list:
                    final_recommend_list.append(ani_id)
                if len(final_recommend_list) >= 10:
                    break
    else:
        print('This user is cold-boot userwill use popluar recommendation')
        final_recommend_list = list()
        for i in top10_anime_list:
            ani_id = i[0]
            if ani_id not in final_recommend_list:
                final_recommend_list.append(int(ani_id))
#     print(final_recommend_list)
    return final_recommend_list
```

## Collaborative Filtering wtih Surprise

```python
## def functions will be used for surprise libarary

def KNN_ItemBased_recommend(item_name,algo, df = df_rate,n_recommendations = 10):
    # Find nearest neighbors of given item using KNN 
    item_id = anime_list[anime_list["Name"] == item_name]["MAL_ID"].values[0]
    inner_id = algo.trainset.to_inner_iid(item_id )
    neighbors = algo.get_neighbors(inner_id, k=n_recommendations)
    anime_neighbors = (algo.trainset.to_raw_iid(id)  for id in neighbors)

    ans = []
    for item in anime_neighbors:
        ans.append(item)
    
    # Create final result dataframe
    final_predictions =  pd.DataFrame(ans, columns=['anime_id'])
    predictions_full = final_predictions.merge(anime_list, left_on = "anime_id", right_on = "MAL_ID")[["anime_id","Name","Score"]]
    return  predictions_full



## User based Recommender, will be used for all alogorithms and be compared
def UserBased_recommend(user_id, algo ,df = df_rate,n_recommendations = 10):
    # Find all anime the user has already rated before
    rated_animes = df.loc[df['user_id']== user_id, 'anime_id']
    unique_ids = df['anime_id'].unique()
    
    # Find all anime ids that the user has not rated 
    animes_to_predict = np.setdiff1d(unique_ids, rated_animes)
    
    # Predict all unrated animes
    my_recs = []
    for iid in animes_to_predict:
        my_recs.append((iid, algo.predict(user_id,iid=iid).est))
    
    #Select top animes from prediction and make final dataframe
    final_predictions =  pd.DataFrame(my_recs, columns=['anime_id', 'predictions']).sort_values('predictions', ascending=False).head(n_recommendations)
    predictions_full = final_predictions.merge(anime_list, left_on = "anime_id", right_on = "MAL_ID")[["anime_id","Name","Score",'predictions']]
    return predictions_full
```

# Neural Network

```python
#Get Train and test data for neural network

# Set target and feature columns
features = ["user_id","anime_id"]
target = "rating"
# Split features and target
X = df_rate[features]
y = df_rate[target]
# Split train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
# Split set into user and anime vector
X_train_array = [X_train.iloc[:, 0], X_train.iloc[:, 1]]
X_test_array = [X_test.iloc[:, 0], X_test.iloc[:, 1]]
```

```python
# Define recommemder net

#Input parameters
n_users = max(df_rate['user_id'].unique()) + 1
n_animes = max(df_rate['anime_id'].unique()) + 1
n_factors = 10

def RecommenderNet(n_users, n_animes, n_factors):
    # Create user embedding, with an extra bias embdding
    user = Input(shape=(1,))
    user_emb = Embedding(n_users, n_factors, embeddings_initializer='he_normal',embeddings_regularizer=l2(1e-6))(user)
    user_emb = Reshape((n_factors,))(user_emb)
    user_bias = Embedding(n_users, 1, embeddings_initializer='he_normal',embeddings_regularizer=l2(1e-6))(user)
    user_bias = Reshape((1,))(user_bias)
  
    # Create anime embedding, with an extra anime embdding
    anime = Input(shape=(1,))
    anime_emb = Embedding(n_animes, n_factors, embeddings_initializer='he_normal',embeddings_regularizer=l2(1e-6))(anime)
    anime_emb = Reshape((n_factors,))(anime_emb)
    anime_bias = Embedding(n_users, 1, embeddings_initializer='he_normal',embeddings_regularizer=l2(1e-6))(anime)
    anime_bias = Reshape((1,))(anime_bias)
    
    #Concate user_emb and anime_emb, add bias, add output layer and create model
    x = Dot(axes=1)([user_emb, anime_emb])
    x = Add()([x, user_bias, anime_bias])
    x = Activation('sigmoid')(x)
    x = Lambda(lambda x: x * 10)(x)
    model = Model(inputs=[user, anime], outputs=x)
    model.compile(loss='mean_squared_error', optimizer= Adam(lr=0.001),metrics=['mae', metrics.RootMeanSquaredError(name='rmse')])
   

    return model
```

```python
# Create model
model = RecommenderNet(n_users, n_animes, n_factors = 10)
model.summary()
```

```python
# Train model
history = model.fit(x=X_train_array, y=y_train, batch_size=64, epochs= 5, verbose=1, validation_data=(X_test_array, y_test))
```

```python
# Recommend animes based on recommender net
def model_recommend(model , user_id, df = df_rate , n_recommendations = 10):
    rated_animes = df.loc[df['user_id']== user_id, 'anime_id']
    unique_ids = df['anime_id'].unique()
    animes_to_predict = list(np.setdiff1d(unique_ids, rated_animes))
    ids = pd.DataFrame([user_id] * len(animes_to_predict))
    animes =  pd.DataFrame( animes_to_predict)
    results =  model.predict([ids, animes])
    my_recs = [(animes_to_predict[i], results[i][0]) for i in range(len(animes_to_predict))]
    
    final_predictions =  pd.DataFrame(my_recs, columns=['anime_id', 'predictions']).sort_values('predictions', ascending=False).head(n_recommendations)
    predictions_full = final_predictions.merge(anime_list, left_on = "anime_id", right_on = "MAL_ID")[["anime_id","Name","Score",'predictions']]
    return predictions_full
```

```python
# Find n_factors with minimum mae and rmse
n_lists = [10,20,40,60,80,100]
min_mae = float("Inf")
min_mae_n = 10
min_rmse = float("Inf")
min_rmse_n = 10
results = []


for n in n_lists:
    # Create model
    model = RecommenderNet(n_users, n_animes, n)
    # Train model
    history = model.fit(x=X_train_array, y=y_train, batch_size=64, epochs= 5, verbose=1, validation_data=(X_test_array, y_test))
    his = history.history
    mae_val = np.mean(his["val_mae"])
    rmse_val = np.mean(his["val_rmse"])
    results.append((n, mae_val, rmse_val))
    if mae_val < min_mae:
        min_mae = mae_val
        min_mae_n = n
    if rmse_val < min_rmse:
        min_rmse = rmse_val
        min_rmse_n = n
        
        
print("min mae is:",min_mae)
print("min mae n_factors is:",min_mae_n)
print("  ")
print("min rmse is:",min_rmse)
print("min rmse n_factors is:",min_rmse_n)
rmse['Nerual Network'] = min_rmse
mae['Nerual Network'] = min_mae
```

```python
columns = ["n","mae","rmse"]
results = pd.DataFrame(results, columns=columns)
ax = results.plot(x="n", y = ["mae","rmse"],kind='line', figsize=(19, 9))
ax.set_title('Evaluation results nerual network recommender with differement n factors', fontsize='x-large')
plt.xticks(rotation=0)
plt.legend(fontsize='x-large')
plt.show()
```
