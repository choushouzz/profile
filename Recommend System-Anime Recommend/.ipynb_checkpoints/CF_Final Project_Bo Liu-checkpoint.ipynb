# Step 4 : 2nd Similarity Matrix Anime-Anime 

```python
# Function: calculate the similarity, list 1 = user's list, list 2 = anime database 
# Algorithm：Cosine Similarity 

def calCosineSimilarity(list1, list2): 
    res = 0
    denominator1 = 0.0
    denominator2 = 0.0
    for i in range(len(list1)):
        val1 = list1[i]
        val2 = list2[i]
        if float(val1) == 0.0 or float(val2) == 0.0:
            continue
        res += (val1 * val2)
        denominator1 += val1 ** 2
        denominator2 += val2 ** 2
    if denominator1 == 0.0 or denominator2 == 0.0:  # when denominator = 0, similarity = 0
        return 0
    else:
        return res / (math.sqrt(denominator1 * denominator2))
```

```python
# Retrieve the cold-boot anime popular list
rank_anime_mapping = list()
for i in anime_all_data[['MAL_ID','Score']].values.tolist():
#     if i[1] == 'Unknown':
#         rank_anime_mapping.append([i[0], 0])
#         print(rank_anime_mapping([i[0], 0]))
#     else:
        rank_anime_mapping.append(i)
top10_anime_list = sorted(rank_anime_mapping, key=lambda x:x[1], reverse=True)[0:10]  
# Sorted by score of each anime
```

```python
# Create an empty anime similarity matrix for further use
animeSimMatrix = np.zeros((len(anime_values), len(anime_values)))
```

```python
animeSimMatrix
```

```python
# if user is in the database，then recommend based on the content similarity，if not then give recomment Top 10
def recommend(user_id):
    if user_id in user_id_list:
        print('Have rated other animations, will use CF recommendation')
        
        # retrieve all the anime ID that he/she has rated and the score ＞ 3
        a_list = anime_list_all_data[(anime_list_all_data['user_id'] == user_id) & 
                                     (anime_list_all_data['rating'] >= 3)]['anime_id']
        print('Anime amount which has been rated and the score is higher than 3：', len(a_list))
        
        if len(a_list) == 0: 
        # which means he/she is not interested in the anime watched before, then recommend from popularity lsit
            final_recommend_list = list()
            print('Have no interested animes before, will recommend from popular list:')
            for i in sorted(rank_anime_mapping, key=lambda x:x[1], reverse=True):
                ani_id = i[0]
                if ani_id not in a_list and ani_id not in final_recommend_list:
                    final_recommend_list.append(ani_id)
                if len(final_recommend_list) >= 10:
                    break
                    
        else:   
            recommend_list = list()
            # caculate anime's similarity one by one
            for a in a_list:
            # loop each anime ID in the watched & rated list 
                a = int(a)
                index = anime_list.index(a)
                start_list = anime_values[int(index)]
                rate_list = list()
                
                for j in range(len(anime_values)):
                    if j == index: 
                    # the anime ID is itself，then do not caculate similarity
                        continue
                        
                    if float(animeSimMatrix[index, j]) > 0.0:
                    # which means already existed in the similarity matrix                     
                        cal1 = animeSimMatrix[index, j]
                        
                    else:
                        cal1 = calCosineSimilarity(start_list, anime_values[j])
                    # caculate the similarity between user's watched anime and other animes                     
                        animeSimMatrix[index, j] = cal1 
                        # BTW, improve the similarity matrix
                        animeSimMatrix[j, index] = animeSimMatrix[index, j]
                    rate_list.append([j, cal1])

                # from user's similarity list，find top 10.
                top10_a_list = list(sorted(rate_list, key=lambda x:x[1], reverse=True)[0:10])
                recommend_list += top10_a_list
  
            final_recommend_list = list()
            print("Find 10 most similar animations： ")
            for i in sorted(recommend_list, key=lambda x:x[1], reverse=True):
                id_ = i[0]
                ani_id = anime_list[id_]
                if ani_id not in final_recommend_list:
                    final_recommend_list.append(ani_id)
                if len(final_recommend_list) >= 10:
                    break
    else:
        print('This user is cold-boot user，will use popluar recommendation：')
        final_recommend_list = list()
        for i in top10_anime_list:
            ani_id = i[0]
            if ani_id not in final_recommend_list:
                final_recommend_list.append(int(ani_id))
#     print(final_recommend_list)
    return final_recommend_list
```

# Step 5: Evaluate with MAE & RSME 

True list : Retrieve 10 animes that user has rated 

Test list : Use top 3 rated score of the true list, then recommend 10 animations

Error : 1 - hit rate（how many succeed/total true amount）

```python
# Select 200 user_ID randomly as test set，caculate MAE value.
import random

all_count = 200
test_user_id_list = random.sample(user_id_list, all_count)
mae_list = list()
rmse_list = list()

for test_uid in test_user_id_list: 
# test each user id
    print(test_uid,'Start Recommend')
    # fisrtly retrieve the anime list that the user has watched and rated before
    # initialize the is_recomm_t flag, 0 means false, 1 means true.
    is_recom_t = 0
    anime_l = list(anime_list_all_data[anime_list_all_data['user_id'] == test_uid]['anime_id'].values)[0:10]
    
    # get user's score rated after 6
    test_anime_l = list(anime_list_all_data[(anime_list_all_data['user_id'] == test_uid) & (anime_list_all_data['rating'] > 6)]['anime_id'].values)
    print('Animation rated over score = 6: ', len(test_anime_l), ' totally')
    
    
    # There is no animation he/she rated over score = 6, then recommend the popular Top 10.
    if len(test_anime_l) < 1:
        pred_list = top10_anime_list 
        
    else:
        # There is 1~3 animations he/she rated over score = 6, take the complete list         
        if len(test_anime_l) > 0 and len(test_anime_l) < 3:
            test_list = test_anime_l
        elif len(test_anime_l) > 3:
            test_list = test_anime_l[0:3]
        else:
            test_list = anime_l[0:3]
        recommend_list = list()
        for a in test_list: 
            # Take the highest 3 animations he/she rated，start recommend 10 animations
            # Retrieve top 10 animations with the highest similarity
            #if the recommendation already existed in the list he/she has watched and rated, then means success.
            a = int(a)
            index = anime_list.index(a)
            start_list = anime_values[int(index)]
            rate_list = list()
            for j in range(len(anime_values)):
                if j == index: 
                    # do not count its similarity
                    continue
                    
                cal1 = calCosineSimilarity(start_list, anime_values[j])
                rate_list.append([j, cal1])
                
            # Retrieve Top 10 highest similarity
            top10_a_list = list(sorted(rate_list, key=lambda x:x[1], reverse=True)[0:10])
            recommend_list += top10_a_list
            
        final_recommend_list = list()
        for i in sorted(recommend_list, key=lambda x:x[1], reverse=True):
            id_ = i[0]
            ani_id = anime_list[id_]
            if ani_id not in final_recommend_list:
                final_recommend_list.append(ani_id)
            if len(final_recommend_list) >= 10:
                break
        pred_list = final_recommend_list
    
    # calculate the true & test lists similarity, then calculate MAE & RSME
    # get sum section of two lists
    try:
        if len(pred_list) > 0:
            bing_list = list(set(pred_list[0:len(anime_list)]) & set(anime_list))
        else:
            bing_list = list()
    except:
        bing_list = list()
#     print(len(bing_list)/ len(anime_l))
    mae_list.append(np.abs(1-(len(bing_list)/ len(anime_l))))
    rmse_list.append(np.square(1-(len(bing_list)/ len(anime_l))))
```

```python
print('The MAE(Mean Absolute Error) is: ',np.mean(mae_list))
print('The RSME(Root Mean Square Error) is: ',np.sqrt(np.mean(rmse_list)))
```
